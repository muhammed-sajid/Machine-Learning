```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()

df = pd.DataFrame(data= data.data, columns= data.feature_names)
print(df.head())
```

#Preprocessing Steps
#Handling Missing Values:#
 The breast cancer dataset from sklearn is clean and does not contain missing values. However, if there were missing values, we could use techniques like imputation (mean, median, or mode) or drop rows/columns with missing values.

#Feature Scaling:#
The features in this dataset have different scales. Scaling is necessary for algorithms like SVM, k-NN, and Logistic Regression, which are sensitive to the magnitude of the features. We use StandardScaler to standardize the features to have a mean of 0 and a standard deviation of 1.

```python
df['target'] = data.target
print(df.head())
```

```python
print("Handling Missing Values:")
print(df.isnull().sum())
```

```python
X = df.drop('target', axis=1)
y = df['target']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
Scaler = StandardScaler()
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.3, random_state=42)
X_train = Scaler.fit_transform(X_train)
X_test = Scaler.transform(X_test)
```

# Logistic Regression


 Logistic Regression is a linear model used for binary classification. It estimates the probability of a sample belonging to a particular class using the logistic function.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
log_reg  = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
# Predict and evaluate
Linear_y_pred = log_reg.predict(X_test)

print("Logistic Regression Accuracy\n")
Log_acc = accuracy_score(y_test, Linear_y_pred)
print("Logistic Accuracy:",Log_acc)
Log_class_report = classification_report(y_test, Linear_y_pred)
print("\nLogistic classification report\n",Log_class_report)
log_conf_matrix  = confusion_matrix(y_test, Linear_y_pred)
print("\nLogistic confusion matrix\n",log_conf_matrix)
```

# Decision Tree Classifier

 A Decision Tree splits the data into subsets based on feature values, creating a tree-like structure. It makes decisions by following the path from the root to a leaf node.

```python
from sklearn.tree import DecisionTreeClassifier

dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)
```

```python
# Predict and evaluate
Decision_y_pred = dt_classifier.predict(X_test)

print("Decision Tree Classifier\n")
Decision_acc = accuracy_score(y_test, Decision_y_pred)
print("Decision Accuracy:",Log_acc)
Decision_class_report = classification_report(y_test, Decision_y_pred)
print("\nDecision classification report\n",Decision_class_report)
Decision_conf_matrix  = confusion_matrix(y_test, Decision_y_pred)
print("\nDecision confusion matrix\n",Decision_conf_matrix)
```

# Random Forest Classifier

 Random Forest is an ensemble method that builds multiple decision trees and combines their outputs to improve generalization and reduce overfitting.

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train, y_train)
```

```python
# Predict and evaluate
Random_y_pred = rf_classifier.predict(X_test)

print("Random Forest Classifier\n")
Random_acc = accuracy_score(y_test, Random_y_pred)
print("Random  Accuracy:",Log_acc)
Random_class_report = classification_report(y_test, Random_y_pred)
print("\nRandom classification report\n",Random_class_report)
Random_conf_matrix  = confusion_matrix(y_test, Random_y_pred)
print("\nRandom confusion matrix\n",Random_conf_matrix)
```

# Support Vector Machine (SVM)

SVM finds the hyperplane that best separates the classes in the feature space. It can use kernel functions to handle non-linear relationships.


```python
from sklearn.svm import SVC

svm = SVC()
svm.fit(X_train, y_train)
```

```python
#predict and evaluate
svm_y_pred = svm.predict(X_test)

print(" Support Vector Machine: \n")
svm_acc = accuracy_score(y_test, svm_y_pred)
print("SVM Accuracy:",svm_acc)
svm_class_report = classification_report(y_test, svm_y_pred)
print("\nSVM Classifiaction report:\n",svm_class_report)
svm_conf_matrix = confusion_matrix(y_test, svm_y_pred)
print("\nSVM confusion matrix:\n", svm_conf_matrix)
```

# k-Nearest Neighbors (k-NN)

k-NN classifies a sample based on the majority class among its k-nearest neighbors in the feature space.


```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
#predict and evaluate
knn_y_pred = knn.predict(X_test)

print(" k-Nearest Neighbors: \n")
knn_acc = accuracy_score(y_test, knn_y_pred)
print("KNN Accuracy:",knn_acc)
knn_class_report = classification_report(y_test, knn_y_pred)
print("\nSVM Classifiaction report:\n",knn_class_report)
knn_conf_matrix = confusion_matrix(y_test, knn_y_pred)
print("\nSVM confusion matrix:\n", knn_conf_matrix)
```

#Create a Comparison Table

```python
def evaluate_classifier(model, model_name, X_train, X_test, y_train, y_test):
  model.fit(X_train, y_train)

  y_pred = model.predict(X_test)

  report = classification_report(y_test, y_pred, output_dict=True)

  metrics = {
      "Algorithm": model_name,
      "Precision (0)" : report["0"]['precision'],
      "Recall (0)" : report["0"]['recall'],
      "F1-Score (0)": report["0"]['f1-score'],
      "Precision (1)" : report["1"]['precision'],
      "Recall (1)" : report["1"]['recall'],
      "F1-Score (1)": report["1"]['f1-score'],
      "Accuracy" : accuracy_score(y_test, y_pred)

  }
  return metrics

classifiers = [
    ("Logistic Regression", LogisticRegression()),
    ("Decision Tree", DecisionTreeClassifier(random_state=42)),
    ("Random Forest", RandomForestClassifier(random_state=42)),
    ("SVM", SVC(kernel='linear', random_state=42)),
    ("k-NN", KNeighborsClassifier(n_neighbors=5))
]
results = []
for name, model in classifiers:
    metrics = evaluate_classifier(model, name, X_train, X_test, y_train, y_test)
    results.append(metrics)

comparison_table = pd.DataFrame(results)

print(comparison_table)
```
