```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
print(data.head())
```

#1. Loading and Preprocessing
python

```python
data.info()
```

```python
print("Handling Missing values:")
print(data.isnull().sum())
```

```python
df = data.drop(columns=['species'], errors='ignore')
print(df.head())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
print(scaled_data)
```

```python
import seaborn as sns

sns.scatterplot(x=df['sepal length (cm)'], y=df['sepal width (cm)'], hue=df['petal length (cm)'], palette='viridis' )
```

#2.Clustering Algorithm Implementation

#A) KMeans Clustering (4 marks)


**KMeans Clustering:**
KMeans is a centroid-based clustering algorithm.

It partitions the dataset into k clusters, where each cluster is represented by its centroid.

The algorithm iteratively minimizes the sum of squared distances between data points and their assigned cluster centroids.

**Why KMeans is suitable **

The Iris dataset has a small number of features and a clear structure, making it suitable for centroid-based clustering.

KMeans works well when the number of clusters (k) is known or can be estimated.

```python

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

KMeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = KMeans.fit_predict(scaled_data)
label = KMeans.labels_
```

```python
label
```

```python
data
```

```python
sns.scatterplot(x=data.iloc[:,0], y=data.iloc[:,1], hue=data['Cluster'], palette='viridis')
plt.title("K-Means Clustering in iris Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal width")
plt.show()
```

#B) Hierarchical Clustering

**Hierarchical Clustering:**

Hierarchical clustering builds a tree-like structure (dendrogram) to represent data points.

It can be agglomerative (bottom-up) or divisive (top-down).

Agglomerative clustering starts with each data point as a single cluster and merges the closest pairs of clusters iteratively

**Why Hierarchical Clustering is suitable for the Iris dataset:**

It provides a dendrogram, which helps visualize the relationships between clusters.

It does not require specifying the number of clusters in advance.

```python
from scipy.cluster.hierarchy import dendrogram,linkage

#Perfrom hierarchy clustering
z = linkage(scaled_data, method='ward')
```

```python
labels = data['sepal length (cm)'].tolist()
print(labels)
```

```python
dendrogram(z, labels=labels, leaf_rotation=90)
plt.title("Hierarchy Clustering Dendrogram in iris Dataset")
plt.xlabel("Data points")
plt.ylabel("Euclidean Distance")
plt.show()
```

#Agglomarative Clustering

```python
from sklearn.cluster import AgglomerativeClustering

#perfrom  AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=3, linkage='ward')
data['Cluster']= hc.fit_predict(scaled_data)
```
